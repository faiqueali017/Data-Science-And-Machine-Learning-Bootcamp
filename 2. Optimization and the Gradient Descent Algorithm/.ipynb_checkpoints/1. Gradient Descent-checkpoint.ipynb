```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### Example: 01 -A simple cost function
#### $$f(x) = x^2 + x +1$$

```python
#Cost function
def f(x):
    return x**2 + x + 1
```

```python
# Make data
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
# Plot
plt.plot(x_1, f(x_1)) #plot(x,y)

# Setting graph
plt.xlim(-3,3)
plt.ylim(0,8)

#Labeling
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16)
plt.title('Cost Function', fontsize = 17)

plt.show()
```

### Slopes and Derivatives
<p>Derivate of the function.</p>

```python
def df(x):
    return 2*x + 1
```

```python
#Plot two side by side
plt.figure(figsize=[15,5]) #[width, height]

# 01: Chart:- Cost Function
plt.subplot(1,2, 1) #(no of rows, no of columns, index of first plot)

# Setting graph
plt.xlim(-3,3)
plt.ylim(0,8)

#Labeling
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16)
plt.title('Cost Function', fontsize = 17)

plt.plot(x_1, f(x_1), color = 'blue', linewidth = 3) #plot(x,y)


# 02: Chart:- Derivative
plt.subplot(1,2, 2) #(no of rows, no of columns, index of second plot)

# Setting graph
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.grid()

plt.xlabel('X', fontsize = 16)
plt.ylabel('df(x)', fontsize = 16)
plt.title('Slope of the cost function', fontsize = 17)

plt.plot(x_1, df(x_1), color = 'skyblue', linewidth = 3) #plot(x,y)

plt.show()
```

### Gradient Descent

```python
new_x = 3 #Starting point
previous_x = 0 #Temp
step_multiplier = 0.1
precision = 0.0001
x_list = [new_x] #Holding x-values
slope_list = [df(new_x)] #Holding slope values

for n in range(500):
    #1- Random guess as initial in temp variable
    previous_x = new_x
    
    #2- Calculate slope of it
    gradient = df(previous_x)
    
    #3- Calculating error (Value of slope at previous value of x)
    #   Our program in learning 
    new_x = previous_x - step_multiplier * gradient
    
    #4- Calculating step size
    step_size = abs(new_x - previous_x)
    
    #5- Append values to list
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    #6- For terminating
    if step_size < precision:
        break
    

print('Local min occurs at:', new_x)
print('Slope of df(x) value at this point:', df(new_x))
print('f(x) value or cost at this point:', f(new_x))
```

### Superimpose gradient descent calculations here

```python
#Plot two side by side
plt.figure(figsize=[20,5]) #[width, height]

'''
# 01: Chart:- Cost Function
'''
plt.subplot(1,3, 1) #(no of rows, no of columns, index of first plot)

# Setting graph
plt.xlim(-3,3)
plt.ylim(0,8)

#Labeling
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16)
plt.title('Cost Function', fontsize = 17)

plt.plot(x_1, f(x_1), color = 'blue', linewidth = 3, alpha = 0.8) #plot(x,y)

#Ploting scatter plot
values = np.array(x_list) #Type conversion
plt.scatter(x_list,f(values), color = 'red', s = 100, alpha=0.6)


'''
# 02: Chart:- Derivative
'''
plt.subplot(1,3, 2) #(no of rows, no of columns, index of second plot)

# Setting graph
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.grid()

plt.xlabel('X', fontsize = 16)
plt.ylabel('df(x)', fontsize = 16)
plt.title('Slope of the cost function', fontsize = 17)

plt.plot(x_1, df(x_1), color = 'skyblue', linewidth = 3, alpha = 0.6) #plot(x,y)
#Ploting scatter plot
plt.scatter(x_list,slope_list, color = 'red', s = 100, alpha=0.6)


'''
# 03: Chart:- Gradient Descent (close up)
'''
plt.subplot(1,3, 3) #(no of rows, no of columns, index of second plot)

# Setting graph
plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)
plt.grid()

plt.xlabel('X', fontsize = 16)
plt.title('Gradient Descent (close up)', fontsize = 17)

plt.plot(x_1, df(x_1), color = 'skyblue', linewidth = 6, alpha = 0.8) #plot(x,y)t
plt.scatter(x_list,slope_list, color = 'red', s = 300, alpha=0.6)

plt.show()
```
