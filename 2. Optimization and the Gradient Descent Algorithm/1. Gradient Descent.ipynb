```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Example: 01 -A simple cost function
### $$f(x) = x^2 + x +1$$

```python
#Cost function
def f(x):
    return x**2 + x + 1
```

```python
# Make data
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
# Plot
plt.plot(x_1, f(x_1)) #plot(x,y)

# Setting graph
plt.xlim(-3,3)
plt.ylim(0,8)

#Labeling
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16)
plt.title('Cost Function', fontsize = 17)

plt.show()
```

### Slopes and Derivatives
<p>Derivate of the function.</p>

```python
def df(x):
    return 2*x + 1
```

```python
#Plot two side by side
plt.figure(figsize=[15,5]) #[width, height]

# 01: Chart:- Cost Function
plt.subplot(1,2, 1) #(no of rows, no of columns, index of first plot)

# Setting graph
plt.xlim(-3,3)
plt.ylim(0,8)

#Labeling
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)', fontsize = 16)
plt.title('Cost Function', fontsize = 17)

plt.plot(x_1, f(x_1), color = 'blue', linewidth = 3) #plot(x,y)


# 02: Chart:- Derivative
plt.subplot(1,2, 2) #(no of rows, no of columns, index of second plot)

# Setting graph
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.grid()

plt.xlabel('X', fontsize = 16)
plt.ylabel('df(x)', fontsize = 16)
plt.title('Slope of the cost function', fontsize = 17)

plt.plot(x_1, df(x_1), color = 'skyblue', linewidth = 3) #plot(x,y)

plt.show()
```

### Gradient Descent

```python
new_x = 3 #Starting point
previous_x = 0 #Temp
step_multiplier = 0.1
precision = 0.0001
x_list = [new_x] #Holding x-values
slope_list = [df(new_x)] #Holding slope values

for n in range(500):
    #1- Random guess as initial in temp variable
    previous_x = new_x
    
    #2- Calculate slope of it
    gradient = df(previous_x)
    
    #3- Calculating error (Value of slope at previous value of x)
    #   Our program in learning 
    new_x = previous_x - step_multiplier * gradient
    
    #4- Calculating step size
    step_size = abs(new_x - previous_x)
    
    #5- Append values to list
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    #6- For terminating
    if step_size < precision:
        break
    

print('Local min occurs at:', new_x)
print('Slope of df(x) value at this point:', df(new_x))
print('f(x) value or cost at this point:', f(new_x))
```

## Example: 02 -Multiple Minima vc Initial Guess & Advanced
### $$g(x) = x^4 - 4x^2 +5$$

```python
#Cost function
def g(x):
    return pow(x,4) - 4*pow(x,2) + 5

#Derivative of g(x)
def dg(x):
    return 4*pow(x,3) - 8*x
```

```python
# Making some data
x_2 = np.linspace(-2,2,1000)
```

```python
#Plot two side by side
plt.figure(figsize=[15,5]) #[width, height]

# 01: Chart:- Cost Function
plt.subplot(1,2, 1) #(no of rows, no of columns, index of first plot)

# Setting graph
plt.xlim(-2,2)
plt.ylim(0.5,5.5)

#Labeling
plt.xlabel('X', fontsize = 16)
plt.ylabel('g(x)', fontsize = 16)
plt.title('Cost Function', fontsize = 17)

plt.plot(x_2, g(x_2), color = 'blue', linewidth = 3) #plot(x,y)


# 02: Chart:- Derivative
plt.subplot(1,2, 2) #(no of rows, no of columns, index of second plot)

# Setting graph
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.grid()

plt.xlabel('X', fontsize = 16)
plt.ylabel('dg(x)', fontsize = 16)
plt.title('Slope of the cost function', fontsize = 17)

plt.plot(x_2, dg(x_2), color = 'skyblue', linewidth = 3) #plot(x,y)

plt.show()
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_iter=300):
    
    new_x = initial_guess #Starting point
    x_list = [new_x] #Holding x-values
    slope_list = [derivative_func(new_x)] #Holding slope values

    for n in range(max_iter):
        #1- Random guess as initial in temp variable
        previous_x = new_x

        #2- Calculate slope of it
        gradient = derivative_func(previous_x)

        #3- Calculating error (Value of slope at previous value of x)
        #   Our program in learning 
        new_x = previous_x - multiplier * gradient

        #4- Calculating step size
        step_size = abs(new_x - previous_x)

        #5- Append values to list
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        #6- For terminating
        if step_size < precision:
            break
            
    #7- Return
    return new_x, x_list, slope_list
```

```python
# Calling func by its all args
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg,
                                                 initial_guess=0.5,
                                                 multiplier=0.02,
                                                 precision=0.001)
print('Local min occurs at:', local_min)
print('No of steps:', len(list_x))
```

```python
# Calling func by its only req args
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg,
                                                 initial_guess=-0.5)
print('Local min occurs at:', local_min)
print('No of steps:', len(list_x))
```

```python
# Calling gradient_descent func
local_min, list_x, deriv_list = gradient_descent(derivative_func=dg,
                                                 initial_guess=0.1)
#Plot two side by side
plt.figure(figsize=[15,5]) #[width, height]

# 01: Chart:- Cost Function
plt.subplot(1,2, 1) #(no of rows, no of columns, index of first plot)

# Setting graph
plt.xlim(-2,2)
plt.ylim(0.5,5.5)

#Labeling
plt.xlabel('X', fontsize = 16)
plt.ylabel('g(x)', fontsize = 16)
plt.title('Cost Function', fontsize = 17)

plt.plot(x_2, g(x_2), color = 'blue', linewidth = 3, alpha = 0.8) #plot(x,y)
plt.scatter(list_x, g(np.array(list_x)),color = 'red', s=100, alpha=0.6)


# 02: Chart:- Derivative
plt.subplot(1,2, 2) #(no of rows, no of columns, index of second plot)

# Setting graph
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.grid()

plt.xlabel('X', fontsize = 16)
plt.ylabel('dg(x)', fontsize = 16)
plt.title('Slope of the cost function', fontsize = 17)

plt.plot(x_2, dg(x_2), color = 'skyblue', linewidth = 5, alpha=0.6) #plot(x,y)
plt.scatter(list_x, deriv_list ,color = 'red', s=100, alpha=0.5)

plt.show()
```

## Example: 03 -Divergence, Overflow and Python Tuples
### $$h(x) = x^5 - 2x^4 +2$$

```python
#Cost function
def h(x):
    return pow(x,5) - 2*pow(x,4) + 2

#Derivative of g(x)
def dh(x):
    return 5*pow(x,4) - 8*pow(x,3)
```

```python
# Make data
x_3 = np.linspace(start=-2.5, stop=2.5,num=1000)
```

```python
# Calling gradient_descent func
local_min, list_x, deriv_list = gradient_descent(derivative_func=dh,
                                                 initial_guess=-0.2, max_iter=71)
#Plot two side by side
plt.figure(figsize=[15,5]) #[width, height]

# 01: Chart:- Cost Function
plt.subplot(1,2, 1) #(no of rows, no of columns, index of first plot)

# Setting graph
plt.xlim(-1.2,2.5)
plt.ylim(-1,4)

#Labeling
plt.xlabel('X', fontsize = 16)
plt.ylabel('h(x)', fontsize = 16)
plt.title('Cost Function', fontsize = 17)

plt.plot(x_3, h(x_3), color = 'blue', linewidth = 3, alpha = 0.8) #plot(x,y)
plt.scatter(list_x, h(np.array(list_x)),color = 'red', s=100, alpha=0.6)


# 02: Chart:- Derivative
plt.subplot(1,2, 2) #(no of rows, no of columns, index of second plot)

# Setting graph
plt.xlim(-1,2)
plt.ylim(-4,5)
plt.grid()

plt.xlabel('X', fontsize = 16)
plt.ylabel('dh(x)', fontsize = 16)
plt.title('Slope of the cost function', fontsize = 17)

plt.plot(x_3, dh(x_3), color = 'skyblue', linewidth = 5, alpha=0.6) #plot(x,y)
plt.scatter(list_x, deriv_list ,color = 'red', s=100, alpha=0.5)

plt.show()

print('Local min occurs at:', local_min)
print('Cost at this minimum is:', h(local_min))
print('No of steps:', len(list_x))
```
